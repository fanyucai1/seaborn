```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import ssl
ssl._create_default_https_context = ssl._create_unverified_context#####导入ssl时关闭证书验证,防止出现certificate verify failed
import warnings
warnings.filterwarnings(action="ignore")
#有一些数据，想“看看”它长什么样，我们一般会画直方图（Histogram）。现在你也可以用核密度估计。
iris = sns.load_dataset("iris")
plt.style.use('ggplot')
```

```python
iris.head()
```

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.kdeplot(iris['sepal_length'])
```

```python
#######2D KDE plots
sns.jointplot(x=iris['sepal_length'],y=iris['sepal_width'],kind='kde')
```
