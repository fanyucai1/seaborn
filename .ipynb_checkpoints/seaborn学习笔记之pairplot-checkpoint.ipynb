```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import ssl
ssl._create_default_https_context = ssl._create_unverified_context#####导入ssl时关闭证书验证,防止出现certificate verify failed
import warnings
warnings.filterwarnings(action="ignore")####忽略无用的警告
plt.style.use('ggplot')####采用ggplot的风格
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.set_style("white")######There are five preset seaborn themes: darkgrid, whitegrid, dark, white, and ticks
g = sns.pairplot(iris,hue='species',kind='reg')###kind : {‘scatter’, ‘reg’}
```
