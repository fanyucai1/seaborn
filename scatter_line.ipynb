```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import ssl
ssl._create_default_https_context = ssl._create_unverified_context#####导入ssl时关闭证书验证,防止出现certificate verify failed
import warnings
warnings.filterwarnings(action="ignore")####忽略无用的警告
plt.style.use('ggplot')####采用ggplot的风格
iris = sns.load_dataset("iris")
iris.head()
```

```python
g=sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species')#https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot
```

```python
g=sns.lineplot(x='sepal_length',y='sepal_width',data=iris,hue='species')#https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot
```

```python
#draw linear regression models
g=sns.lmplot(x='sepal_length',y='sepal_width',data=iris,hue='species')#https://seaborn.pydata.org/tutorial/regression.html#regression-tutorial
```

```python
g=sns.lmplot(x='sepal_length',y='sepal_width',data=iris,col='species')#https://seaborn.pydata.org/tutorial/regression.html#regression-tutorial
```

```python
#Plotting a regression in other contexts
sns.jointplot(x='sepal_length',y='sepal_width',data=iris, kind="reg");#https://seaborn.pydata.org/tutorial/regression.html#regression-tutorial
```
